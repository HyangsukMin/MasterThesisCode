```python
def main(N):
    # DP !!
    d = [0 for _ in range(N+1)]
    for i in range(2, N + 1):
        d[i] = d[i-1] + 1
        if i % 3 == 0:
            d[i] = min(d[i], d[i//3] + 1)
        if i % 2 == 0:
            d[i] = min(d[i], d[i//2] + 1)
    return d[N]
N = int(input())
print(main(N))
```
